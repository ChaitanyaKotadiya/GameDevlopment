# Q.12. Write a program to demonstrate how to detect when the left mouse button is clicked and display the coordinates of the mouse click.

```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui

# Initialize global variables
width = 400
height = 400
position = ""

# Mouse click handler
def click(pos):
    global position
    position = "Mouse Clicked at (x={}, y={})".format(pos[0], pos[1])

# Draw handler
def draw(canvas):
    canvas.draw_text("Click anywhere on the canvas", [50, 50], 20, "White")
    canvas.draw_text(position, [50, 100], 18, "White")

# Create the frame
frame = simplegui.create_frame("Mouse Click Coordinates", width, height)

# Register event handlers
frame.set_draw_handler(draw)
frame.set_mouseclick_handler(click)

# Start the frame
frame.start()
```

After clicking on canvas it will show the coordinates of mouse.

# Q. 16. Write a program to load an image and displays it on the canvas.
(For example:

```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui

# Constants for the canvas size
CANVAS_WIDTH = 400
CANVAS_HEIGHT = 300

# Load the image (provide the correct path or use a local image)
image = simplegui.load_image("https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSHkqyHvkSF2gaBLVx6FZvDNNtxnJ49cxmz2E9A1K5q&s")

# Event handler to draw the image on the canvas
def draw(canvas):
    canvas.draw_image(image,
                      (image.get_width() / 2, image.get_height() / 2),  # Center of the image
                      (image.get_width(), image.get_height()),  # Size of the image
                      (CANVAS_WIDTH / 2, CANVAS_HEIGHT / 2),  # Position on the canvas
                      (CANVAS_WIDTH, CANVAS_HEIGHT))  # Size on the canvas

# Create the frame
frame = simplegui.create_frame("Image Display", CANVAS_WIDTH, CANVAS_HEIGHT)

# Register the draw handler
frame.set_draw_handler(draw)

# Start the frame
frame.start()
```

# Q.17. Write a program to Animate image using timer.

```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
# Constants for the canvas size and image dimensions
CANVAS_WIDTH = 400
CANVAS_HEIGHT = 300
FRAME_WIDTH = 100
FRAME_HEIGHT = 100

# Load the image (provide the correct path or use a local image)
image = simplegui.load_image("https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTS_P3z6mq60lUDnN6iVJ9YENrp2Ur424qIww&usqp=CAU")

# Current frame index
frame_index = 0

# Timer interval in milliseconds
timer_interval = 100  # Adjust this value to control the animation speed

# Timer event handler
def update_frame():
    global frame_index
    frame_index = (frame_index + 1) % (image.get_width() // FRAME_WIDTH)

# Event handler to draw the image on the canvas
def draw(canvas):
    canvas.draw_image(image,
                      (FRAME_WIDTH / 2 + frame_index * FRAME_WIDTH, FRAME_HEIGHT / 2),  # Center of the current frame
                      (FRAME_WIDTH, FRAME_HEIGHT),  # Size of the frame
                      (CANVAS_WIDTH / 2, CANVAS_HEIGHT / 2),  # Position on the canvas
                      (CANVAS_WIDTH, CANVAS_HEIGHT))  # Size on the canvas

# Create the frame
frame = simplegui.create_frame("Image Animation", CANVAS_WIDTH, CANVAS_HEIGHT)

# Register the draw handler
frame.set_draw_handler(draw)

# Create a timer with the specified interval and register the timer event handler
timer = simplegui.create_timer(timer_interval, update_frame)

# Start the timer
timer.start()

# Start the frame
frame.start()
```

# Q. 20. Write a program to add Game Over sound.

```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import random

# List of mystical answers
mystical_answers = [
    "Yes, definitely.",
    "It is decidedly so.",
    "Without a doubt.",
    "Yes, definitely.",
    "You may rely on it.",
    "As I see it, yes.",
    "Most likely.",
    "Outlook good.",
    "Yes.",
    "Signs point to yes.",
    "Reply hazy, try again.",
    "Ask again later.",
    "Better not tell you now.",
    "Cannot predict now.",
    "Concentrate and ask again.",
    "Don't count on it.",
    "My reply is no.",
    "My sources say no.",
    "Outlook not so good.",
    "Very doubtful."
]

# Load simulated sound
magic_sound = simplegui.load_sound("https://file-examples.com/storage/fee3d1095964bab199aee29/2017/11/file_example_WAV_1MG.wav")

# Function to get a mystical answer
def get_mystical_answer():
    return random.choice(mystical_answers)

# Event handler for input field
def input_handler(text):
    print("Your Question: " + text)
    answer = get_mystical_answer()
    print("Mystical Octosphere says:", answer)
    magic_sound.rewind()
    magic_sound.play()

# Create the frame
frame = simplegui.create_frame("Mystical Octosphere", 400, 200)

# Create the input field
input_field = frame.add_input("Enter your yes-or-no question:", input_handler, 200)

# Start the frame
frame.start()
```

After writing yes or no it will start the sound
